```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('labels.csv')
df
```

```python
cols = list(df.columns)[6:]
for col in cols:
    d = dict(df[col].value_counts())
    d[np.NaN] = len(df) - sum(d.values())
    plt.bar(list(map(lambda x: str(x), list(d.keys()))), height = list(d.values()))
    plt.title("Visualization of label {} frequency".format(col))
    plt.ylabel("Frequency")
    plt.xlabel("Value")
    print(d)
    plt.show()
```
